#**Descripcion del hardware utilizado para el entrenamiento de algoritmos de AI**

##**CPU (Central Processing Unit)**

La cpu es un elemento de vital importancia para cualquier dispositivo que podamos encontrar en el mercado como (Computadores, Celulares, Tablets,etc.) incluso hasta en relojes, se le denomina el cerebro de cualquier dispositivo.

de manera mas tecnica la CPU es un procesador de proposito general, que se encarga de ejecutar diferentes tipos de aplicaciones, como calculos matematicos,  video juegos, reproducior musica, editar texto, navegar por internet, etc.

los componentes de una CPU son:

*   **(ALU) Unidad Logica Aritmetica**:
La ALU realiza operaciones aritmeticas y logicas las cuales se ejecutan una a una y no de manera paralela.

<img src="https://www.researchgate.net/profile/Saumyakanta-Sarangi/publication/279757619/figure/fig2/AS:669640208416778@1536666075391/Figure1-Block-diagram-of-a-4-bit-ALU.ppm">

*   **(CU) Unidad de Control**:
La CU dirige el tráfico de información entre los registros de la CPU y conecta con la ALU las instrucciones extraídas de la memoria.
*   **Registros Internos**:
no accesibles (de instrucción, de bus de datos y bus de dirección) y accesibles de uso específico (contador programa, puntero pila, acumulador, flags, etc.) o de uso general

La CPU se encarga de procesar abosultamente todas las señales emitidas, tanto internas como externas, desde aplicaciones y programas del sistema hasta la comunicación con perifericos como altavoces, raton, teclado, usbs, etc.

##**GPU (Graphics Processing Unit)** 

La GPU es el corazon, cerebro y es el componente esencial  de todas las tarjetas graficas conocidas, al igual que la CPU, la GPU realiza todas las operaciones aritmeticas para el procesamiento de imagenes, primeramente pensada para juegos pero con mucho potencial para el desarrollo de algoritmos de inteligencia artificial.

La GPU cuenta con una arquitectura en paralelo enorme que consiste de mieles de núcleos más pequeños y eficaces, y que se diseñaron para resolver varias tareas al mismo tiempo.

**Las GPU tienen miles de núcleos para procesar cargas de trabajo en paralelo de forma eficiente.**

<img src="https://images.nvidia.com/content/nvem/docs/IO/144271/cpu-and-gpu.jpg">

Para que tengas una idea el GPU a pesar de que puede ser considerado un coprocesador se diferencia del procesador central en su arquitectura que está orientada al trabajo y manejo de gráficos, siendo incluso mucho más eficiente que el procesador para esa tarea, pero muy incipiente para realizar otras tareas, es por esta razón que nunca podrá reemplazar al procesador central dentro de un equipo de cómputo avanzado como un ordenador, servidor, etc.

##**TPU (Tensor Processing Unit)**

La TPU es un circuito integrado de aplicación desarrollado especificamente para procesar tareas de inteligencia artificial de manera efectiva, otorgando una arquitectura muy robusta para el procesamiento de grandes sets de datos. Desarrollado por **Google** especificamente para el aprendizaje profundo con redes neuronales artificiales haciendo uso optimizado de la libreria TensorFlow.

En comparación con las unidades de procesamiento grafico, estas unidades se diseñaron para manejar un mayor volumen de calculo de precisión reducida(desde 8 bits de precisión). En Fotos, una TPU individual puede procesar más de 100 millones de fotos al día y es actualmente una de las tecnologias mas potentes para el procesamiento de estos algoritmos.
